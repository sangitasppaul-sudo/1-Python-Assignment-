# **5. What is mutability in Python?**
  - Mutability means whether an object’s value can be changed (modified) after it is created. In Python, some data types are mutable (can change), while others are immutable (cannot change).
  **Mutable Objects** - Values can be changed without changing the object’s identity (memory address). Examples: list, dict, set, bytearray numbers = [1, 2, 3] print(id(numbers)) # memory address before numbers.append(4) # modify list print(numbers) # [1, 2, 3, 4] print(id(numbers)) # same memory address (mutable). **Immutable Objects** Once created, values cannot be changed. Any “change” creates a new object with a different memory address. Examples: int, float, string, tuple, frozenset, bool x = "hello" print(id(x)) # memory address before x = x + " world" # creates a new string print(x) # hello world print(id(x)) # different memory address (immutable)

# **6. Why are lists mutable, but tuples are immutable?**
  - Lists are designed to store collections of items that may change. You can add, remove, or modify elements without creating a new list. Internally, Python gives lists a dynamic memory structure that allows resizing and modification. Example - fruits = ["apple", "banana", "cherry"] fruits[1] = "mango" # modify element fruits.append("grape") # add element print(fruits) # ['apple', 'mango', 'cherry', 'grape'] **Tuples in Python are immutable because** - 1. Tuples are designed to represent fixed collections of items. 2. Once created, their values cannot change (no add, remove, or update). 3. Internally, Python fixes the memory allocation of a tuple, making it safe for hashing (so tuples can be used as keys in dictionaries, unlike lists).Example: colors = ("red", "green", "blue") colors[1] = "yellow" # Error: 'tuple' object does not support item assignment Key Reason for the Difference • Lists → meant for dynamic data that changes. • Tuples → meant for static data that stays fixed. This design choice helps Python balance flexibility (lists) and safety/performance (tuples).

# **13. What is the difference between for and while loops?**
  - Difference Between for loop and while loop:
  **1.Initialization**
  for loop: Initialization is usually done in the loop itself (like iterating over a sequence or range).
  while loop: Initialization must be done separately before the loop starts.
  **2.Condition**
  for loop: Works with a sequence (range, list, string, etc.) and stops automatically when the sequence ends.
  while loop: Runs until a given condition becomes False.
  **3. Iteration Control**
  for loop: Automatically handles increment/decrement when iterating through a sequence.
  while loop: Programmer must manually update the loop variable to avoid infinite loops.
  **4. Usage**
  for loop: Best when the number of iterations is known.
  while loop: Best when the number of iterations is not known in advance and depends on a condition.
  **5. Readability**
  for loop: More concise and easier to read for definite iteration.
  while loop: Can become longer since initialization, condition, and increment are separate.
  **6. Infinite Loop Possibility**
  for loop: Less prone to infinite loops (unless break is misused).
  while loop: More prone to infinite loops if the condition never becomes False.

# **14. Describe a scenario where a while loop is more suitable than a for loop.**
  - Scenario where a while loop is more suitable than a for loop:
  A while loop is more suitable when the number of iterations is not known in advance and the loop should continue running until a certain condition is met.
  For example, consider a situation where a program repeatedly asks a user to enter a password until they enter the correct one. In this case, we don’t know how many attempts the user will take. A for loop is not ideal here because it usually runs for a fixed number of iterations. Instead, a while loop is best since it continues until the condition (correct password entered) becomes true.
  **Explanation:**
  for loop works well when the number of repetitions is predetermined (like printing numbers from 1 to 10).
  while loop is better when the repetitions depend on a condition that may change during program execution (like waiting for user input, reading data until end-of-file, or monitoring a sensor until it reaches a threshold).

```python
# 14. Write a program to reverse a string using a while loop.
text = input("Enter a string: ")

reversed_text = ""
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a non-negative integer: "))

print("Invalid input! Please enter an integer.")

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    counter = num

while counter > 0:
        factorial *= counter
        counter -= 1

print(f"The factorial of {num} is: {factorial}")
```
